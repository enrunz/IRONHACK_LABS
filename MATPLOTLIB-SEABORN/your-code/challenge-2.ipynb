#  USING SEABORN IS A MUCH BETTER OPTION THAN PLT THIS TIME SO PLEASE FIND SEABORN FIRST AND THEN PLT

```python
# Driver 1 enter codes here

def specs(x, **kwargs):
    
    ax = sns.histplot(x=x,bins=10,alpha=0.5)
    ax.axvline(x.mean(), color='green', lw=1)
    ax.axvline(x.median(), color='red', ls='--', lw=2)

multi=sns.FacetGrid(fitbit, row='Work_or_Weekend',col='Months_encoded',margin_titles=True)
multi.map(specs,'Steps')



```

# NOW THIS IS THE MEAN BY MONTH

```python
by_month=fitbit.pivot_table(index=['Months'],values='Steps')
fig2=plt.figure(figsize=(12,4))
plt.bar(by_month.index,by_month['Steps'],alpha=0.1)
plt.savefig('fig2.png')
```

# AND THE MEDIAN

```python
y=fitbit.groupby('Months').median()['Steps']
fig3=plt.figure(figsize=(12,4))
plt.bar(y.index,by_month['Steps'],alpha=0.1)
plt.savefig('fig3.png')
```

# AND HERE YOU CAN SEE THE DIFFERENCE BETWEEN WEEK AND WEEKEND BY MONTH

```python
w_w_m=fitbit.pivot_table(index=['Months','Work_or_Weekend'],values='Steps').reset_index()

fig4=plt.figure(figsize=(12,4))
plt.bar(w_w_m['Months'],w_w_m['Steps'],alpha=0.1)
plt.savefig('fig4.png')

```

# AND FINALLY JUST SEE THE GLOBAL DIFFERENCE BTW WEEK AND WEEKEND AVGS

```python
w_o_w=fitbit.groupby('Work_or_Weekend').mean()['Steps'].reset_index()

fig5=plt.figure(figsize=(12,4))
plt.barh( w_o_w['Work_or_Weekend'],w_o_w['Steps'],)
plt.yticks([0, 1], ['Weekend', 'Work'])
plt.savefig('fig5.png')
```
