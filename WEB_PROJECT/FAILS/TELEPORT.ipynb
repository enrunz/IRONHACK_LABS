# WEB SCRAPING

How does wikipedia talk about the challenges of a city?
take the three lowest scores and see if they are addressed at all in the wiki page.
also check for bikes. 



```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
SD_low_scores=['travel','housing','commute']
```

```python
ws_url='https://en.wikipedia.org/wiki/San_Diego'
page=requests.get(ws_url).content
page
```

```python
soupSD= BeautifulSoup(page,'html')
sections= [i.text.replace('[edit]','').lower() for i in soupSD.find_all('h2')]

sections
    
```

```python
for i in SD_low_scores:
    if i in sections:
        print(f'{i} is addressed as a section')
    else:
        print(f'{i} is not a main section. We will keep searching')
```

```python
sub_sections= [i.text.replace('[edit]','').strip().lower() for i in soupSD.find_all('h3')]
sub_sections
```

```python
for i in SD_low_scores:
    if i in sub_sections:
        print(f'{i} is addressed as a sub-section')
    else:
        print(f'{i} is not a sub-section. Hmmmmm. We will keep searching')
```

```python
parragraphs= [i.text.replace('[edit]','').strip().lower() for i in soupSD.find_all('p')]
parragraphs
```

# I COULD CLEAN THIS UP

```python
split_parragraphs=[]
for i in parragraphs:
    y=i.split(' ')
    split_parragraphs.append(y)
split_parragraphs
```

# I COULD COUNT THE NUMBER OF APPEARANCES

```python
for i in SD_low_scores:
    if i in split_parragraphs:
        print(f'{i} is mentioned in a parragraph')
    else:
        print(f'{i} is not mentioned in a parragraph. SOMEONE IS FAKING IT')
```
