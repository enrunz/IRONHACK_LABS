#### Select all the customers who have bought more than 50 items ?

```python
# your answer here
visits=orders.groupby('CustomerID').count().reset_index()[['CustomerID','InvoiceNo']]
visits.rename(columns={'InvoiceNo':'Visits'},inplace=True)
visits.sort_values('Visits',ascending=False)
visits[(visits['Visits']>50)]
```

#### Select orders from Spain that are above 50 items

```python
# your answer here
orders[(orders['Country']=='Spain')&(orders['Quantity']>50)]
```

#### Select all free orders

```python
# your answer here
orders[(orders['Description'].str.contains('lunch bag'))]
```

#### Select all orders made in the month of August

```python
# your answer here
orders[(orders['month']== 8)]
```

#### Select how many orders are made by countries in the month of August
##### Hint: Use value_counts()

```python
orders_country_month=orders.groupby(['Country'])['month'].value_counts()
ocm=pd.DataFrame(orders_country_month)
ocm.rename(columns={'month':'no. transactions'},inplace=True)
ocm.reset_index(inplace=True)
ocm[(ocm['month']==8)][['Country','no. transactions']]
```

#### What's the  average amount of money spent by country
