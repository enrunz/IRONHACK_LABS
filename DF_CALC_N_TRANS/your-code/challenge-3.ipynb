We break down the main problem into several sub problems:

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
cus=orders.groupby('CustomerID').sum()['amount_spent']
customers=pd.DataFrame(cus)
customers
```

```python
cus.quantile(.95)
```

```python
cus.quantile(.75)
```

```python
customers['Type of customer']=np.where((cus>cus.quantile(.75))&(cus<=cus.quantile(.95)),'Preferred',
                                       (np.where((cus>cus.quantile(.95)),'VIP','Normal')))
customers
```

# Q2: How to identify which country has the most VIP Customers?



Provide your solution for Q2 below:

```python
customer_country=orders.drop_duplicates(subset=['CustomerID'])
customer_by_country=customer_country[['CustomerID','Country']]
y=customer_by_country.sort_values(['CustomerID'])
y
```

```python
customers['country']=y['Country'].values
customers
```

```python
df1=customers.groupby(['country','Type of customer']).count().reset_index()
df1
```

```python
datos_filtrados=df1[(df1['Type of customer'])=='VIP'].sort_values('amount_spent',ascending=False)
datos_filtrados.rename(columns={'amount_spent':'Number of clients VIP'},inplace=True)
datos_filtrados
```

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
datos_filtrados2=df1[(df1['Type of customer'])!='Normal']
VIP_Preferred=datos_filtrados2.groupby('country').sum()
VIP_Preferred.rename(columns={'amount_spent':'Number of clients VIP+Preferred'},inplace=True)
VIP_Preferred.sort_values('Number of clients VIP+Preferred',ascending=False)

```
